```python
# Image Path in my drive
image_path = '/content/drive/MyDrive/AF/PIC.jpg'

img = cv2.imread(image_path)

print(f"Image shape: {img.shape} -(height, width, channels)")
#RGB value for one pixel
pixel_bgr = img[112, 112,0:3]
print(f"Pixel at ({112}, {112}) → BGR values: {pixel_bgr}") #BGR value for that pixel

b, g, r = pixel_bgr
print(f"R = {r}, G = {g}, B = {b}")
```

```python
# Show the img
plt.imshow(img)
plt.title("PIC.jpg")
plt.axis('off')
plt.show()
```

```python
# Show the img in RGB
# img_rgb1 = img[:, :, ::-1]   # using reverse [B, G, R] → [R, G, B]

img_rgb= img.copy()
img_rgb[:,:,0] = img[:,:,2]
img_rgb[:,:,1] = img[:,:,1]
img_rgb[:,:,2] = img[:,:,0]

plt.imshow(img_rgb)
plt.title("Manually Converted in Original")
plt.axis('off')
plt.show()
```

```python
b,g,r = img[:,:,0],img[:,:,1],img[:,:,2]
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1) #Blue Channel
plt.imshow(b)
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(1, 3, 2) #Green Channel
plt.imshow(g)
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1, 3, 3) #Red Channel
plt.imshow(r)
plt.title("Red Channel")
plt.axis('off')

plt.show()
```
